# Read info file

```python
snapNum = 41
fname = f"./info_{snapNum:05d}.txt"
```

```python
with open(fname, "r") as file:
    lines = file.readlines()
```

```python
lines
```

```python
info = {}

with open(fname, "r") as file:
    for line in file:
       
        if "=" in line:
            key, value = line.split("=")
            key = key.strip()     # remove whitespace
            value = value.strip() # remove whitespace
            
            try:
                if "." in value:
                    value = float(value)
                
                else:
                    value = int(value)
            
            except ValueError:
                pass # leave the value as a string
                
            info[key] = value
            
```
